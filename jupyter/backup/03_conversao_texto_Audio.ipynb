```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import torch
```

```python
pipeline = KPipeline(lang_code='p')
```

```python
path_texto = '/home/elton/Projetos/Translater_Video/jupyter/temp/resultado_teste.txt'
texto = ""

audio_speaker = "/home/elton/Projetos/Translater_Video/audio_prompt/base_audio_1.wav"

with open(path_texto, 'r', encoding='utf-8') as f:
    texto = f.read()
    
path_temp = '/home/elton/Projetos/Translater_Video/jupyter/temp/'+'a'
```

```python
generator = pipeline(
    texto, voice='af_heart', # <= change voice here
    speed=1, split_pattern=r'\n+'
)
```

```python
for i, (gs, ps, audio) in enumerate(generator):
    print(i)  # i => index
    print(gs) # gs => graphemes/text
    print(ps) # ps => phonemes
    display(Audio(data=audio, rate=24000, autoplay=i==0))
    samplerate = 24000
    sf.write(f'{path_temp}_{i}.wav', samplerate, audio) # save each audio file
```
